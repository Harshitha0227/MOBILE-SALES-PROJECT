```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv('mobile_sales.csv')

# Clean and prepare data
data.columns = data.columns.str.strip()
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')
data['UnitsSold'] = pd.to_numeric(data['UnitsSold'], errors='coerce')
data['Date'] = pd.to_datetime(data['Date'].str.strip(), dayfirst=True, errors='coerce')

# Create total sales column
data['Total_Sales'] = data['Price'] * data['UnitsSold']

# Group by date
daily_sales = data.groupby('Date')['Total_Sales'].sum()

# Plot
plt.figure(figsize=(10,5))
daily_sales.plot(kind='line', marker='o', color='teal', title='Daily Sales Trend')
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('mobile_sales.csv')

# Convert Date column properly
df["Date"] = pd.to_datetime(df["Date"].str.strip(), dayfirst=True, errors="coerce")

# Remove rows with invalid dates
df = df.dropna(subset=["Date"])

# Compute average price per day
avg_price_per_day = df.groupby("Date")["Price"].mean().reset_index()

# Plot
plt.figure(figsize=(10,5))
sns.lineplot(x="Date", y="Price", data=avg_price_per_day, marker="o", color="teal")
plt.title("Average Price Over Time", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Average Price")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read dataset
data = pd.read_csv('mobile_sales.csv')

# Ensure 'Price' is numeric
data["Price"] = pd.to_numeric(data["Price"], errors="coerce")
data = data.dropna(subset=["Price"])

# Plot histogram
plt.figure(figsize=(8,5))
plt.hist(data["Price"], bins=15, edgecolor='black', color='darkblue')
plt.title("Price Distribution", fontsize=14)
plt.xlabel("Price", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
#barplot on top 5 location by revenue
top_locations = data.groupby('Location')['TotalRevenue'].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,5))
top_locations.plot(kind='bar', color='orange')
plt.title('Top 5 Locations by Revenue')
plt.xlabel('Location')
plt.ylabel('TotalRevenue')
plt.tight_layout()
plt.show()
```

```python
# Count payment methods
payment_counts = data['PaymentMethod'].value_counts()

# Plot
plt.figure(figsize=(6,6))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Set3.colors)
plt.title("Payment Method Distribution")
plt.show()
```

```python
#piechart on sales share by brand
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("mobile_sales.csv")
df['TotalRevenue'] = pd.to_numeric(df['TotalRevenue'], errors='coerce')

brand_sales = df.groupby('Brand')['TotalRevenue'].sum().nlargest(10)

plt.pie(brand_sales, labels=brand_sales.index, autopct='%1.1f%%', startangle=90)
plt.title("Sales Share by Brand")
plt.show()
```
